### On average, how much money each company has received at Series A round?

```python
# First let's convert the currencies to USD$

exchange_rates = {
    '€': 1.1,   # 1 Euro = 1.1 USD
    '£': 1.3,   # 1 GBP = 1.3 USD
    '¥': 0.007, # 1 Yen = 0.007 USD
    '₹': 0.012  # 1 INR = 0.012 USD
}

def clean_money_raised(value):
    value_str = str(value)
    cleaned_value = ''.join(filter(str.isdigit, value_str))
    
    if cleaned_value:
        numeric_value = float(cleaned_value)
        
        # Check for currency symbols and convert if necessary
        for symbol, rate in exchange_rates.items():
            if symbol in value_str:
                return numeric_value * rate
        
        return numeric_value
    else:
        return None

# Apply the function to the MONEY_RAISED column
investment_df['MONEY_RAISED'] = investment_df['MONEY_RAISED'].apply(clean_money_raised)

# Filter for Series A funding type and calculate the average money raised
series_a_df = investment_df[investment_df['FUNDING_TYPE'] == 'Series A']
avg_money_raised = series_a_df['MONEY_RAISED'].dropna().mean()
print(f"Average money raised in Series A: ${avg_money_raised:,.2f}")
```

##### The average funding raised in Series A rounds is an indicator of the capital available for companies at a critical stage of growth and expansion.


### What is the average money raised by funding type?

```python
# Now that we have converted, the money to USD$, let's remove  currency symbols and other non-numeric characters for consistent analysis

investment_df['MONEY_RAISED_CLEAN'] = investment_df['MONEY_RAISED'].replace(
    {'\$': '', ',': '', '€': '', '£': '', 'CN¥': '', '—': '', '¥': ''}, regex=True
).apply(pd.to_numeric, errors='coerce')

# Now let's try the average money raised plot
average_money_raised = investment_df.groupby('FUNDING_TYPE')['MONEY_RAISED_CLEAN'].mean().dropna()

plt.figure(figsize=(10, 6))
average_money_raised.plot(kind='bar')
plt.title("Average Money Raised by Funding Type")
plt.xlabel("Funding Type")
plt.ylabel("Average Money Raised ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### What is the average funding raised by companies in each industry category?


```python
# Let's groupby and calculate the average funding by category

def clean_money_raised(money_raised):
  try:
    if isinstance(money_raised, (int, float)):
      return money_raised

    cleaned_money_raised = money_raised.replace('$', '').replace(',', '').replace('—', '')

# Let's Convert to float
    return float(cleaned_money_raised)
  except (TypeError, ValueError, AttributeError):
    return float('nan')

# Let's apply the cleaning function to the MONEY_RAISED column
investment_df['MONEY_RAISED'] = investment_df['MONEY_RAISED'].apply(clean_money_raised)

avg_funding_by_category = investment_df.merge(company_df, on='COMPANY_NAME').groupby('CATEGORY')['MONEY_RAISED'].mean()
avg_funding_by_category.head()
```

##### By examining the average funding raised by category, we can identify industries that attract larger investments and potentially higher growth opportunities.

## <span style="color:red"> **Geography**</span>

### What are the top locations for companies by revenue range and employee count



```python
# Group by location for revenue and employee data
top_locations_by_revenue = company_df.groupby('LOCATION')['REVENU_RANGE'].value_counts().unstack().fillna(0)
top_locations_by_employees = company_df.groupby('LOCATION')['EMPLOYEE_NUMBER'].value_counts().unstack().fillna(0)

# Reset index to make 'LOCATION' a column and calculate total employee count per location
top_locations_by_revenue = top_locations_by_revenue.reset_index()
top_locations_by_employees['EMPLOYEE_COUNT'] = top_locations_by_employees.sum(axis=1)
top_locations_by_employees = top_locations_by_employees.reset_index()

# Let's then merge the dataframes on 'LOCATION' and select the top 10 locations by total employee count
merged_data = pd.merge(top_locations_by_revenue, top_locations_by_employees[['LOCATION', 'EMPLOYEE_COUNT']], on='LOCATION', how='inner')
top_10_locations = merged_data.nlargest(10, 'EMPLOYEE_COUNT')

top_10_locations.set_index('LOCATION').plot(kind='bar', stacked=True, figsize=(12, 8), edgecolor='black')
plt.title("Top 10 Locations by Revenue Range and Employee Count")
plt.xlabel("Location")
plt.ylabel("Counts")
plt.xticks(rotation=45)
plt.legend(title="Revenue Ranges and Total Employee Count", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

##### This highlights the geographical distribution of companies based on revenue and employee count, offering insights into regional economic activity and talent pools.
